### 前処理

```python
train_copy = train.copy()
test_copy = test.copy()
```

```python
# yearカラムの修正
train_copy = dp.fix_year_column(train_copy)
test_copy = dp.fix_year_column(test_copy)

# odometerカラムの修正
train_copy = dp.fix_odometer_column(train_copy)
test_copy = dp.fix_odometer_column(test_copy)
```

```python
train_copy.isnull().sum()
```

```python
# 最頻値で埋める
dp.fillna_with_mode(train_copy, ['fuel', 'title_status', 'type'])
dp.fillna_with_mode(test_copy, ['fuel', 'title_status', 'type'])
        
# stateの欠損値をregionから導いて埋める
dp.fill_missing_state(train_copy, area_mapping)
dp.fill_missing_state(test_copy, area_mapping)
```

```python
train_copy.isnull().sum()
```

```python
# manufacturerカラムを整形
dp.normalize_manufacturer_column(train_copy)
dp.normalize_manufacturer_column(test_copy)
```

```python
train_copy['manufacturer'].value_counts()
```

```python
#製造メーカのデータ確認
manufacturer_values_count = train_copy["manufacturer"].nunique()
print("値の種類の数:", manufacturer_values_count)

manufacturer_grouped = train_copy.groupby("manufacturer")
for name, group in manufacturer_grouped:
    print(name)
```

```python
# 新車のデータが入ってる？→中古車販売店でも新車はあるので問題ない
train_copy['condition'].value_counts()
```

```python
# sizeカラムの値も揃ってなさそう
train_copy['size'].value_counts()
```

```python
# sizeカラムの整形
train_copy = dp.normalize_size_column(train_copy)
test_copy = dp.normalize_size_column(test_copy)
```

```python
from pandas.core.api import DataFrame
import seaborn as sns

# データ整形後にカテゴリ変数とpriceとの相関を確認
categorical_columns = ['region', 'manufacturer', 'condition', 'cylinders', 'fuel', 'title_status',
                       'transmission', 'drive', 'size', 'type', 'paint_color', 'state']

for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    top_categories = train_copy[col].value_counts().index[:5]
    sns.boxplot(x=col, y='price', data=train_copy[train_copy[col].isin(top_categories)])
    plt.title(f'Boxplot of price against {col}')
    plt.xticks(rotation=90)
    plt.show()
```

```python
train_copy.head()
```

```python
# conditionカラムのダミー変数化
train_copy = dp.encode_condition_column(train_copy)
test_copy = dp.encode_condition_column(test_copy)

# ワンホットに変換
train_copy = pd.get_dummies(train_copy, columns=["manufacturer", "size", "state", "region"])
test_copy = pd.get_dummies(test_copy, columns=["manufacturer", "size", "state", "region"])
```

```python
# trainデータとtestデータのカラムを揃える
missing_cols_train = set(train_copy.columns) - set(test_copy.columns)
for col in missing_cols_train:
    test_copy[col] = 0

missing_cols_test = set(test_copy.columns) - set(train_copy.columns)
for col in missing_cols_test:
    train_copy[col] = 0

# trainデータとtestデータのカラムを並び替える（順序を揃える）
test_copy = test_copy[train_copy.columns]
```

```python
test_copy.info()
```

### データの分割

```python
#priceは予測対象で学習に必要なため別途targetの変数に格納する
target = train_copy["price"]
#学習させる特徴以外を削除
train_set = train_copy.drop(columns=['id', 'year', 'cylinders', 'fuel', 'title_status', 'transmission', 'drive', 'type', 'paint_color', "price"], axis=1)
test_set = test_copy.drop(columns=['id', 'year', 'cylinders', 'fuel', 'title_status', 'transmission', 'drive', 'type', 'paint_color', "price"], axis=1)
```

```python
train_set.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(train_set, target, random_state = 82)
print(X_train.shape, X_valid.shape, y_train.shape, y_valid.shape)
```

### 学習と検証

```python
#ランダムフォレストで学習
model = RandomForestRegressor(random_state=100)
model.fit(X_train, y_train)
```

```python
#予測
pred = model.predict(X_valid)
```

```python
#評価
score = mean_absolute_percentage_error(y_valid, pred)
print(score*100)
```

```python
#横軸が予測結果、縦軸が正解価格
plt.scatter(pred, y_valid)
plt.xlabel("predict")
plt.ylabel("price")
plt.show()
```
